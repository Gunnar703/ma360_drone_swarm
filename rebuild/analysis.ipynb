```python
from swarm_model import SwarmModel
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

from numba import jit
```

```python
model = SwarmModel()
```

```python
TRIALS_PER_NUM = 10

n_drones = list(range(1, 101))
returns  = []

pbar = tqdm(total = len(n_drones) * TRIALS_PER_NUM)

@jit
def calc():
    for number in n_drones:
        trials = np.zeros(TRIALS_PER_NUM)
        for i in range(TRIALS_PER_NUM):
            trials[i] = model.run(number)
            pbar.update()
        returns.append(trials.mean())

calc()
```

```python
plt.plot(n_drones[:len(returns)], returns)
plt.title("Utility vs. Number of Drones")
plt.xlabel("Number of Drones")
plt.ylabel("Utitlity (USD)")
plt.savefig("plot.png")
np.savetxt("returns.txt", np.hstack( (np.array(n_drones[:len(returns)]).reshape(-1,1), np.array(returns).reshape(-1,1)) ))
```
